## The Dataset is Highly Unbalanced

```python
#0 -------> Legit Data
#1 -------> Fradulent Data
```

```python
legit = pd[pd.Class==0]
fraud = pd[pd.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
pd.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
import pandas as pd
```

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = new_dataset.drop(columns='Class', axis=1)
```

```python
Y = new_dataset['Class']
```

```python
print(X.head())
```

```python
print(Y.head())
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Model Selection and Trainning

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

## Model Evaluation Accuracy Score

```python
X_train_pridiction = model.predict(X_train)
trining_data_accuracy = accuracy_score(X_train_pridiction,Y_train)
```

```python
print('Accuracy on Training data :',trining_data_accuracy)
```

## Accuracy on Test Data

```python
X_test_prediction = model.predict(X_test)
test_accuracy_score = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Test Data :',test_accuracy_score)
```
